**Customer Segmentation Using Unsupervised Learning**

**Introduction**
In this task, we aim to develop and train an unsupervised machine learning model to segment customers based on their transactional behavior and other relevant features from the dataset. The goal is to identify customer groups that share similar behaviors and use these segments to discuss strategies for offering coupons to increase loyalty and satisfaction.

**Feature Selection**
We will utilize the following customer demographic and transactional features for segmentation:

**Demographic features:**
* Gender
* City
**Transactional features:**
Coupon usage frequency
Transaction status
Average transaction value
Frequency of transactions
Model Development
We will train and evaluate a K-Means clustering model for customer segmentation. K-Means is a popular unsupervised learning algorithm that partitions the data into K clusters based on their similarities.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the dataset
def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"Error loading data: {e}")

# Preprocess the data
def preprocess_data(data):
    # Check for missing values
    if data.isnull().values.any():
        print("Missing values found. Filling with mean for numerical columns and mode for categorical columns.")
        for col in data.columns:
            if data[col].dtype == 'object':
                data[col] = data[col].fillna(data[col].mode()[0])
            else:
                data[col] = data[col].fillna(data[col].mean())

    # Encode categorical variables
    categorical_cols = data.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])

    # Scale numerical variables
    numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
    scaler = StandardScaler()
    data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

    return data

# Select relevant features
def select_features(data):
    try:

        relevant_features = ['gender', 'city', 'coupon_usage_frequency', 'transaction_status', 'avg_transaction_value', 'transaction_frequency']
        available_features = [col for col in relevant_features if col in data.columns]

        if not available_features:
            print("Desired features not found. Using all numerical features for clustering.")
            available_features = data.select_dtypes(include=['number']).columns.tolist()

        features = data[available_features]
        return features
    except Exception as e:
        print(f"Error selecting features: {e}")
        return None


# Train K-Means model
def train_kmeans(features, k):
    try:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(features)
        return kmeans
    except Exception as e:
        print(f"Error training K-Means model: {e}")
        return None

# Evaluate K-Means model
def evaluate_kmeans(features, kmeans):
    try:
        silhouette = silhouette_score(features, kmeans.labels_)
        inertia = kmeans.inertia_
        return silhouette, inertia
    except Exception as e:
        print(f"Error evaluating K-Means model: {e}")
        return None, None

# Plot inertia and silhouette scores
def plot_scores(inertia_values, silhouette_values):
    try:
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.plot(range(2, 11), inertia_values)
        plt.xlabel('Number of Clusters')
        plt.ylabel('Inertia')
        plt.title('Inertia vs Number of Clusters')
        plt.subplot(1, 2, 2)
        plt.plot(range(2, 11), silhouette_values)
        plt.xlabel('Number of Clusters')
        plt.ylabel('Silhouette Score')
        plt.title('Silhouette Score vs Number of Clusters')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error plotting scores: {e}")

# Main function
def main():
    df = pd.read_excel('E-commerce_data (1).xlsx')

    if df is not None:
        df = preprocess_data(df)
        features = select_features(df)
        if features is not None:
            inertia_values = []
            silhouette_values = []
            for k in range(2, 11):
                kmeans = train_kmeans(features, k)
                if kmeans is not None:
                    silhouette, inertia = evaluate_kmeans(features, kmeans)
                    if silhouette is not None and inertia is not None:
                        inertia_values.append(inertia)
                        silhouette_values.append(silhouette)
            plot_scores(inertia_values, silhouette_values)
            optimal_k = silhouette_values.index(max(silhouette_values)) + 2
            kmeans = train_kmeans(features, optimal_k)
            if kmeans is not None:
                silhouette, inertia = evaluate_kmeans(features, kmeans)
                if silhouette is not None and inertia is not None:
                    print(f'Optimal Number of Clusters: {optimal_k}')
                    print(f'Silhouette Score: {silhouette:.3f}')
                    print(f'Inertia: {inertia:.3f}')

if __name__ == "__main__":
    main()
```

## Feature engineering

### Subtask:
Engineer new features from the existing data in the dataframe `df`.


**Reasoning**:
Load the transaction data and coupon usage data, calculate the new features, and merge them with the main dataframe.



```python
import pandas as pd

# Load transaction data
try:
    transactions = pd.read_csv('transactions.csv')
except FileNotFoundError:
    print("Error: 'transactions.csv' not found.")
    transactions = None
except Exception as e:
    print(f"An error occurred while loading transactions.csv: {e}")
    transactions = None

# Load coupon usage data
try:
    coupon_usage = pd.read_csv('coupon_usage.csv')
except FileNotFoundError:
    print("Error: 'coupon_usage.csv' not found.")
    coupon_usage = None
except Exception as e:
    print(f"An error occurred while loading coupon_usage.csv: {e}")
    coupon_usage = None

if transactions is not None and coupon_usage is not None:
    # Calculate total spending per customer
    total_spending_per_customer = transactions.groupby('customer_id')['transaction_amount'].sum().reset_index()
    total_spending_per_customer.rename(columns={'transaction_amount': 'total_spending'}, inplace=True)
    df = pd.merge(df, total_spending_per_customer, on='customer_id', how='left')

    # Calculate average transaction value per customer
    avg_transaction_value = transactions.groupby('customer_id')['transaction_amount'].mean().reset_index()
    avg_transaction_value.rename(columns={'transaction_amount': 'avg_transaction_value'}, inplace=True)
    df = pd.merge(df, avg_transaction_value, on='customer_id', how='left')

    # Calculate coupon usage frequency
    coupon_usage_frequency = coupon_usage.groupby('customer_id')['coupon_id'].count().reset_index()
    coupon_usage_frequency.rename(columns={'coupon_id': 'coupon_usage_frequency'}, inplace=True)
    df = pd.merge(df, coupon_usage_frequency, on='customer_id', how='left')

    # Select relevant features
    df_selected_features = df[['gender_id', 'city_id', 'total_spending', 'avg_transaction_value', 'coupon_usage_frequency']]
    display(df_selected_features.head())
else:
    print("Could not complete feature engineering due to missing data files.")
```
